```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("TestConnection") \
    .master("local") \
    .getOrCreate()

# Test query
result = spark.sql("SELECT current_timestamp() as timestamp, 'Spark is working!' as message")
result.show()

print("✅ Spark is connected!")
```

```python
# Check if Iceberg catalog is available
try:
    spark.sql("SHOW DATABASES").show()
    print("✅ Iceberg catalog connected to PostgreSQL!")
except Exception as e:
    print(f"❌ Error: {e}")
```

```python
from kafka import KafkaProducer, KafkaConsumer
import json

# Test Kafka connection
try:
    producer = KafkaProducer(
        bootstrap_servers=['localhost:9092'],
        value_serializer=lambda v: json.dumps(v).encode('utf-8')
    )
    
    # Send test message
    producer.send('test-topic', {'message': 'Hello Kafka!', 'status': 'working'})
    producer.flush()
    
    print("✅ Kafka is connected!")
except Exception as e:
    print(f"❌ Kafka Error: {e}")
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("IcebergTest") \
    .master("spark://host.docker.internal:7078") \
    .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions") \
    .config("spark.sql.catalog.iceberg", "org.apache.iceberg.spark.SparkCatalog") \
    .config("spark.sql.catalog.iceberg.type", "jdbc") \
    .config("spark.sql.catalog.iceberg.uri", "jdbc:postgresql://localhost:5432/iceberg_catalog") \
    .config("spark.sql.catalog.iceberg.jdbc.user", "lakehouse") \
    .config("spark.sql.catalog.iceberg.jdbc.password", "lakehouse") \
    .config("spark.sql.catalog.iceberg.warehouse", "s3a://lakehouse/warehouse") \
    .config("spark.hadoop.fs.s3a.endpoint", "http://localhost:8333") \
    .config("spark.hadoop.fs.s3a.access.key", "lakehouse") \
    .config("spark.hadoop.fs.s3a.secret.key", "lakehouse") \
    .config("spark.hadoop.fs.s3a.path.style.access", "true") \
    .getOrCreate()

# Test Iceberg connection
try:
    spark.sql("SHOW DATABASES").show()
    print("✅ Iceberg catalog connected to PostgreSQL!")
except Exception as e:
    print(f"❌ Error: {e}")
```
